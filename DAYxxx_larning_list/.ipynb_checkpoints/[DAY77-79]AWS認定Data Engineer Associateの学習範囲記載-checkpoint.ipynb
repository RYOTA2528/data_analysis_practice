### 模擬試験でのポイント
- (**データ圧縮方式について**)
  - Snappyは列志向ストレージ(Parquet,ORC)と組み合わせることで処理速度も向上
  - 圧縮速度も速く、データ分析用途に適してる。
  - データのストリーミングや分析でリアルタイム処理が求められる場合に使用

- (**フォーマットについて**)
 - Arvo:シリアル化とでシリアル化、スキーマ駆動型で、スキーマ定義がデータと一緒に保存されるため、言語やプラットフォーム間のデータ転送に最適
 - Parquet:ビッグデータ分析に最適化されたカラム型ストレージフォーマット。効果的なデータ圧縮と符号化を提供。Hadoopエコシステムとの統合や複雑なネストデータ構造サポート。分散処理用途に活用。
 - ORC: Parquet同様、カラム型ストレージフォーマットで圧縮とパフォーマンスに優れ、Hadoopエコシステム内で使用。

- (**列志向型データについて**)
 - 列志向型データは分析に適しており、AWSの分析ツールでも列志向でデータを保持することがベストプラクティス
 - （補足）列志向型データ（カラムナデータ）は、データベースやデータウェアハウスでのデータ格納のアーキテクチャの一つです。これに対して、行志向型データベース（ロウストレージ）は、一般的に各レコード（行）を一緒に格納します。列志向型では、データが各列ごとに格納されるため、特定のカラムに対するアクセスや分析において優れた性能を発揮します。

 - （**列志向型の特徴4**）
   - 効率的な圧縮: 列志向型は同じデータ型が連続して格納されるため、圧縮が非常に効率的です。例えば、数値や日付などの繰り返しの多いデータは圧縮され、ストレージ容量を大幅に削減することができます。

   - 特定の列の高速アクセス: 列志向型では、クエリで特定のカラムのみを参照することが多いため、データベースが不要な列を読み込む必要がなくなり、ディスクI/Oを最小化できます。これにより、特に集計やフィルタリング操作が高速化されます。

   - 分析処理に最適: 列志向型データベースは、データ分析、特に大規模なデータセットに対する集計処理や検索処理に適しています。例えば、列ごとに大きな集計処理を行う場合（平均値や合計値など）は、行志向型よりも高速に処理できます。

   - 並列処理の有効性: 列志向型はデータが列ごとに分かれて格納されているため、並列処理や分散処理が非常に有効です。分散型の処理が行いやすく、大規模データの分析に向いています。

### 各種リソース模擬試験でのポイント
- (**Amazon Kinesis Data Streams**)<br>
  - リアルタイムデータの収集、処理をするサービスで、「ログ、イベント、市場価格フィード、ソーシャルメディアフード、IoT、デバイスデータ」等の媒体への処理を得意とする<br>
  - 高スケーラビリティ(サーバレス)、複数のAZへのレプリケート、他サービスとの統合が可能
  - データを送信する側：「プロデューサー」、受信する側：「コンシューマー」
  - プロデューサーからKDSに送信されたレコードは「データロブ」・「パーティションキー」をもつ。パーティションキーに基づいて「シャード、ストリームデータの単位」に分割される

-(**Amazon Managed Service for Apache Flink**)<br>
  - 大量のデータをリアルタイムで処理するユースケースに適してる
  - Kinesisデータストリームから送信されたログデータが特定の閾値を超えた場合にアラートを送信するといったユースケースにも対応
  - フルマネージドサービス
