```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset("titanic")
df
```

```python
plt.hist(df["fare"], bins=40, range=(0, 100)) #binsで粒度を細かく、rangeで範囲を0～100までに絞る
plt.show()
```

```python
# 折れ線グラフのデータの範囲調整・ラベル追加
plt.figure(figsize = (20,5))
plt.plot(df["fare"])
plt.title("test")
plt.xlabel("data_index")
plt.ylabel("farel")
plt.show()
```

```python
# 性別ごとの運賃の表示
df["fare"].hist(by=df["sex"])
plt.show()
```

```python
# 複数のグラフを作成
fig, axes = plt.subplots(2, 2, figsize=(20,10)) #2*2のグラフ
axes[0][0].hist(df["age"], bins=20)
axes[0][1].hist(df["fare"], bins=20)
axes[0][1].set_xlim(0,200) #range=(0,100)と同義
axes[1][0].hist(df["sex"], bins=20)
axes[1][1].hist(df["pclass"], bins=20)
plt.show()
```

```python
# 棒グラフ
# df.groupby("pclass").count()["survived"]
# df.groupby("pclass").count()["survived"].index #groupby後のindex数を表示可能
plt.bar(df.groupby("pclass").count()["survived"],df.groupby("pclass").count()["survived"])
```

```python
# 散布図
plt.scatter(df["fare"],df["age"], alpha=0.2)
plt.show()
```
