```python
# 数値型の列のみを選択/変数に格納
df_numeric = df.select_dtypes(include='number')  # 数値型の列だけを選択
df_numeric = df.select_dtypes(include='number')

# 数値型の列のみでグループ化後の平均を計算
result_mean = df.groupby("survived")[df_numeric.columns].mean()
print(result_mean)
```
