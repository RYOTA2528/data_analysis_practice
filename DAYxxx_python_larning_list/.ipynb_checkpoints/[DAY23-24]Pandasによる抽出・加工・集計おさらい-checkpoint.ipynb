```python
# 数値型の列のみを選択/変数に格納
df_numeric = df.select_dtypes(include='number')  # 数値型の列だけを選択

# 数値型の列のみでグループ化後の平均を計算
result_mean = df.groupby("survived")[df_numeric.columns].mean()
print(result_mean)
```

```python
print(df.dtypes)
```

```python
df.sort_values("age", ascending=False) #ascending=Falseで降順にできる
```

```python
df.isnull().sum() #欠損値の数がわかる
```

```python
df.dropna() #一つでも欠損してれば削除
```

```python
df["age"].head(10)
```

```python
df["age"].fillna(0).head(10) #fillna(0)で欠損値を0に補完
```

```python
df["age"].fillna(df["age"].mean()).head(10) #fillna()で平均の値で補完する処理
```

```python
pd.get_dummies(df) #カテゴリの変数に対して0,1で値を設定してくれる
```

```python
# df.corr()
# 数値型の列のみ選択
df_numeric = df.select_dtypes(include=['number'])
# 数値型の列のみ相関を示す
correlation_matrix = df_numeric.corr()
print(correlation_matrix)
```
