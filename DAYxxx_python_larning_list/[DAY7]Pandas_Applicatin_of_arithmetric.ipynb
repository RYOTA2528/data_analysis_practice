```python
import pandas as pd
import numpy as np
import plotly.express as px
```

```python
df = px.data.gapminder()
```

```python
df
```

```python
#Bool値を表示
df['year'] == 1952
```

```python
#これでTrueの条件のデータフレームを抽出できる
df[df['year'] == 1952]
```

```python
#'gdpPercap'が中央値よりも高くかつyearが2000年よりも後のものを抽出
gdp_per_cap_med = df['gdpPercap'].median() #中央値
condition1 = df['gdpPercap'] > gdp_per_cap_med #中央値の値を超えてる場合
condition2 = df['year']>= 2000
df[condition1 & condition2]
```

```python
#query()を使えばより簡単に抽出可能
df.query('year >= 2000')
```

```python
df.query('(year >= 2000) & (continent == "Asia")')
```

```python
df['continent'].unique()
```

```python
#uniqueな値どれかに属する場合は真となるクエリ
df.query('continent in ["Asia", "Europe", "Africa"]')
```

```python
target_continents =  ["Asia", "Europe", "Africa"]
#@変数名にて変数を使う
df.query('continent in @target_continents')
```

```python
#いずれかを満たすときに真偽値を出す
target_continents =  ["Asia", "Europe", "Africa"]
df['continent'].isin(target_continents)
```

```python
#Trueのもののみのデータフレームを抽出
df[df['continent'].isin(target_continents)]
```

```python
#カラムの中で'A'を含むものが含まれるならその真偽値を返す
df['country'].str.contains('A')
```

```python
df['pop'].describe()
```

```python
pop_med = 7.023596e+06
#lambda関数は無名関数でlambda 引数: 戻り値、apply()は関数の結果をテーブルに適用する
#xはpopの値のこと
df['pop_cat'] = df['pop'].apply(lambda x: 'gt_med' if x >= pop_med else 'lt_med')
```

```python
df
```

```python
df['Log10_pop'] = df['pop'].apply(lambda x: np.log10(x))
#上記で対数log10（X）とは１０回かけ合わせるとXになるかという計算式
```

```python
df
```

```python
def my_func(x):
  return np.log10(x)
```

```python
df['pop'].apply(my_func)
```

```python
def add_continent_code(x):
  if x == 'Asia':
    return 1
  elif x == 'Europe':
    return 2
  else:
    return 3
```

```python
df['Boolvalue'] = df['continent'].apply(add_continent_code)
df
```

```python
#groupby()を用いてcontinent(地域)ごとに平均値の集計をする
df.groupby('continent')['pop'].mean()
```

```python
#get_groupby()でcontinent(地域)の一部の抽出
continent_grp = df.groupby('continent')
continent_grp.get_group('Asia')
```

```python
continent_grp['pop'].mean()
```

```python
df.groupby('continent')['pop'].median().reset_index()
```

```python
#複数の集約(キーは一つ)かつそれぞれのカラムに対して集計方法を変える方法
df.groupby('continent')[['pop', 'gdpPercap']].agg({'pop': 'median', 'gdpPercap': 'mean'})
```

```python
#複数の集約(キーが2つ)かつそれぞれのカラムに対して集計方法を変える方法
df.groupby(['continent', 'year'])['pop'].mean'()
```
